## Loading Required Libaries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
```

## Load the Dataset

```python
Anemia_df=pd.read_csv("diagnosed_cbc_data_v4.csv")
Anemia_df.head()
```

## Data Cleaning

This section ensure the data is in the right format by inspecting the data structure & missing values.

```python
#Inspect the data structure
Anemia_df.info()
```

```python
#Count the number of missing values in each column
Anemia_df.isnull().sum()
```

## Exploratory Data Analysis

```python
Anemia_df.describe()
```

```python
for i,column in enumerate(Anemia_df.columns[:-1],1):
    plt.figure(figsize=(20,15))
    sns.boxplot(x="Diagnosis",y=column,data=Anemia_df)
    plt.title(f"Box Plot Of {column} By Anemia Type")
    plt.tight_layout()
    plt.show()
```

```python
Diagnosis_df=Anemia_df.groupby('Diagnosis').size().reset_index(name='frequency')
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x='Diagnosis',y='frequency',data=Diagnosis_df)
plt.title('Distribution of Anemia type')
plt.xlabel('Anemia type')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
X=Anemia_df.drop(columns=["Diagnosis"])
y=Anemia_df["Diagnosis"]
```

```python
x_corr=X.corr()
plt.figure(figsize=(20,15))
sns.heatmap(x_corr, annot=True, fmt='.2f')
plt.show
```

## Standardizing The Variables 

```python
Scaler=StandardScaler()
X_scaled=Scaler.fit_transform(X)
X_scaled
```

## Dimension Reduction

```python
#PCA
pca=PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate the cumulative explained variance ratio
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Determine the number of components that explain at least 85% of the variance
num_components = np.argmax(cumulative_explained_variance >= 0.85) + 1

print(f'Number of components that explain at least 85% of the variance: {num_components}')
```

```python
# Visualize the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_explained_variance, marker='o', linestyle='--')
plt.axhline(y=0.85, color='r', linestyle='-')
plt.axvline(x=num_components-1, color='b', linestyle='-')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
# Perform PCA with the optimal number of components
pca_optimal = PCA(n_components=num_components)
X_pca_optimal = pca_optimal.fit_transform(X_scaled)

# Create a DataFrame with the optimal PCA results
pca_df_optimal = pd.DataFrame(data=X_pca_optimal, columns=[f'PC{i+1}' for i in range(num_components)])
pca_df_optimal['Diagnosis'] = y
```

```python
# Explained variance by each component
explained_variance = pca_optimal.explained_variance_ratio_
print(f'Explained variance by each component: {explained_variance}')
print(f'Cumulative explained variance: {cumulative_explained_variance[num_components-1]}')
```

## Modeling

```python
# Spliting the data into test and train
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

### Support Vector Machine

```python
# Define the model
svm=SVC(kernel = "linear", random_state=42)
svm.fit(X_train,y_train)
```

```python
# Model Evaluation
svm_y_pred=svm.predict(X_test)
svm_accuracy_score=accuracy_score(y_test,svm_y_pred)
svm_precision_score=precision_score(y_test,svm_y_pred,average='weighted')
svm_f1_score=f1_score(y_test,svm_y_pred,average='weighted')
svm_recall_score=recall_score(y_test,svm_y_pred,average='weighted')
```

```python
svm_accuracy_score
```

### KNN Model

```python
# Define the Model
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
# Model Evaluation
knn_y_pred=knn.predict(X_test)
knn_accuracy_score=accuracy_score(y_test,knn_y_pred)
knn_precision_score=precision_score(y_test,knn_y_pred,average='weighted')
knn_f1_score=f1_score(y_test,knn_y_pred,average='weighted')
knn_recall_score=recall_score(y_test,knn_y_pred,average='weighted')
```

```python
knn_accuracy_score
```

### Random Forest

```python
# Define the Model
rfm=RandomForestClassifier(n_estimators=200,random_state=42,criterion='entropy')
rfm.fit(X_train,y_train)
```

```python
# Model Evaluation
rfm_y_pred=rfm.predict(X_test)
rfm_accuracy_score=accuracy_score(y_test,rfm_y_pred)
rfm_precision_score=precision_score(y_test,rfm_y_pred,average='weighted')
rfm_f1_score=f1_score(y_test,rfm_y_pred,average='weighted')
rfm_recall_score=recall_score(y_test,rfm_y_pred,average='weighted')
```

```python
rfm_accuracy_score
```

```python
#Classification report matrics
print (classification_report(y_test,rfm_y_pred))
```
